```python
"""
Matplotlib:
This assignment is for visualization using matplotlib:
data to use:
url=https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv
titanic = pd.read_csv(url)
Charts to plot:
1. Create a pie chart presenting the male/female proportion
2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender
"""
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_table('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv',delimiter=',')
```

```python
df.head()
```

 pie chart presenting the male/female proportion

```python
# getting number of total males and females
male,female = df['sex'].value_counts()

# plottig pie chart
plt.figure(figsize=(7,7))
plt.pie([male,female],labels=['# of Male','# of Females'],autopct='%1.1f%%', textprops={'fontsize': 20},colors=['#6699ff','#bb99ff'])
plt.title('PieChart representing Male Female Ratio',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(df['age'],df['fare'],hue=df['sex'],palette='seismic',s=150)
plt.title('Scaterplot representing Fare Paid by different age groups differed by gender',fontsize=20)
```
